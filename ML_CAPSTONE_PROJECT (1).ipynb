3. Please justify your data analysis approach.

solution:

Logistic regression is a statistical method used for binary classification, predicting the probability of an observation belonging to a particular class. While logistic regression has its strengths, it may not always be universally better than other models. The suitability of a model depends on the specific characteristics of the data and the goals of the analysis. Here are some advantages of logistic regression that might make it preferable in certain situations:

1. Interpretability:

Logistic regression provides coefficients that can be interpreted as the log-odds of the outcome. This makes it easy to explain the impact of each variable on the probability of the event occurring.

2. Simple and computationally efficient:

Logistic regression is computationally efficient and doesn't require high computational resources compared to more complex models. It is a relatively simple algorithm that can be implemented quickly.

3. No assumptions about data distribution:

Logistic regression doesn't assume a normal distribution of the independent variables. It works well even if the assumptions of linear regression, such as normality of errors, are not met.

4. Well-suited for linearly separable data:

When the decision boundary is approximately linear, logistic regression can perform well. It's effective when the relationship between the independent variables and the log-odds of the dependent variable is roughly linear.

Till here just the data exploration has been done.Now let us move towards the next section of our model that is data preprocessing or data cleaning.
First we remove the unnecessary features/columns from our data that are actually not involved in the data prediction.

**To perform the label encoding technique on the target variable.**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['label_corona']=le.fit_transform(df['Corona'])
df
```

**Performing dummy rule in order to convert categorical data to numerical.**

```python
dum_var=pd.get_dummies(df['Age_60_above'])
df_res=pd.concat([df,dum_var],axis=1)
df_res
```

```python
df_res.drop('No',axis=1,inplace=True)
df_res
```

```python
dum_var=pd.get_dummies(df['Sex'])
df_res1=pd.concat([df_res,dum_var],axis=1)
df_res1
```

```python
df_res1.drop('male',axis=1,inplace=True)
df_res1
```

```python
dum_var=pd.get_dummies(df['Fever'])
df_res2=pd.concat([df_res1,dum_var],axis=1)
df_res2.drop(False,axis=1,inplace=True)
df_res2
```

```python
dum_var=pd.get_dummies(df['Sore_throat'])
df_res3=pd.concat([df_res2,dum_var],axis=1)
df_res3.drop(True,axis=1,inplace=True)
df_res3
```

```python
dum_var=pd.get_dummies(df['Cough_symptoms'])
df_res4=pd.concat([df_res3,dum_var],axis=1)
df_res4.drop('FALSE',axis=1,inplace=True)
df_res4
```

```python
dum_var=pd.get_dummies(df['Shortness_of_breath'])
df_res5=pd.concat([df_res4,dum_var],axis=1)
df_res5.drop('TRUE',axis=1,inplace=True)
df_res5
```

```python
dum_var=pd.get_dummies(df['Headache'])
df_res6=pd.concat([df_res5,dum_var],axis=1)
df_res6.drop('None',axis=1,inplace=True)
df_res6
```

```python
df_res6.columns
```

**To find the shape of the dataframe.**

```python
df_res6.shape
```

```python
df_res6['label_corona'].shape
```

**splitting the data into two parts that is training data and testing data.**

```python
from sklearn.model_selection import train_test_split
#x=df_res5['FALSE'],df_res5['TRUE'],df_res5[True],df_res5[False],df_res5['label_age'],df_res5['label_sex']
y=df_res6['label_corona']
x=df_res6.drop(['Cough_symptoms','Fever','Sore_throat','Shortness_of_breath','Headache','Corona',
               'Age_60_above','Sex','label_corona'],axis=1)
x
y
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

**Creating the model.**

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
x_train.columns=x_train.columns.astype(str)
x_test.columns=x_test.columns.astype(str)
model.fit(x_train,y_train)
```

```python
x_test.head()
```

**Predicting the data**

```python
pred_prb=model.predict_proba(x_test)
pred_prb

```

```python
pred_prb_df=pd.DataFrame(pred_prb[:, :2],columns=['positive','negative'])
pred_prb_df.head()
```

```python
pred_final=model.predict(x_test)
pred_prb_df["model class function"]=pred_final
pred_prb_df["actual"]=y_test
```

**To check whether the model is performing good or bad-CONFUSION MATRIX**

```python
#print(pred_prb_df.isna().sum())
#print(pred_prb_df['actual'].isna().sum())
pred_prb_df['actual']=pred_prb_df['actual'].fillna(0)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(pred_prb_df["actual"],pred_prb_df["model class function"]))
```
